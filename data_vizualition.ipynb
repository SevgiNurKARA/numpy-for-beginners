```python

import numpy as np

import matplotlib.pyplot as plt
##basic intro
```

```python
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)

plt.scatter(x,y)
plt.show()
```

```python
x= np.arange(10)
y = np.random.uniform(0,10,10)
plt.bar(x,y)
```

```python
z = np.random.uniform(0,1,(8,8))
plt.imshow(z)
```

```python
plt.contourf(z)
```

```python
z = np.random.uniform(0,1,4)
plt.legend()
plt.pie(z)
```

```python
z = np.random.normal(0,1,100)
x= np.arange(15)
plt.hist(z,x)
```

```python
x = np.arange(5)
y = np.random.uniform(0,1,5)
plt.errorbar(x,y,y/4)
```

```python
z = np.random.normal(0,1,(100,3))
plt.boxplot(z)
```

```python
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)

plt.scatter(x,y,color="Black")
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.scatter(x,y, linestyle="--",color="Green",marker="*")
plt.title("Trying somethign new")
plt.xlabel("Axis-x")
plt.ylabel("Axis-y")
plt.show()
```

```python
x = np.linspace(0,10,100)
y1,y2 = np.sin(x), np.cos(x)
plt.plot(x,y1,x,y2)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.scatter(x,y, linestyle="--",color="Green",marker="*",lw=2)
plt.title("Trying somethign new")
plt.xlabel("Axis-x")
plt.ylabel("Axis-y")
plt.show()
```

```python
x = np.random.randn(1000,4)
cmap=plt.get_cmap("pink")
colors=cmap([0.2, 0.4, 0.6, 0.8])
plt.hist(x,2,histype="bar",color=colors)
```
